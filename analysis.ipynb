```python
import os
import numpy as np
```

```python
# Set the path to the directory containing the files you want to load
path_to_files = "./data"

# Initialize an empty dictionary to store the loaded data
data_dict = {}

# Loop over all the files in the directory
for filename in os.listdir(path_to_files):
    # Load the data from the file into a NumPy array
    filepath = os.path.join(path_to_files, filename)
    data = np.loadtxt(filepath)
    
    # Add the data array to the dictionary with the filename as the key
    data_dict[filename] = data
```
